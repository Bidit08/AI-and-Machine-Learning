```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

#load image
img = Image.open('image.jpg')

#show image
img.show()

print("image mode: ", img.mode)
print("image size: ", img.size)
```

```python
#converting image to numpy Array
img_array = np.array(img)

# Show image
plt.imshow(img_array)
plt.axis("off")   # remove axis
plt.show()

print("Shape: ", img_array.shape)
print("Data type: ",img_array.dtype)
```

```python
#Basic Operations using NumPy

#Grayscale
gray = np.mean(img_array, axis=2)
gray = gray.astype(np.uint8)

gray_img = Image.fromarray(gray)
gray_img.show()

# Show image
plt.imshow(gray_img)
plt.axis("off")
plt.show()
```

```python
#Invert image
inverted = 255 - img_array

inv_img = Image.fromarray(inverted)
inv_img.show()
# Show image
plt.imshow(inv_img)
plt.axis("off")   # remove axis
plt.show()
```

```python
#Extracting Red channel
red_channel = img_array[:, :, 0]

red_img = Image.fromarray(red_channel)
red_img.show()
# Show image
plt.imshow(red_img)
plt.axis("off")   # remove axis
plt.show()
```

```python
#Cropping Image
cropped = img_array[50:200, 50:200]

crop_img = Image.fromarray(cropped)
crop_img.show()
# Show image
plt.imshow(crop_img)
plt.axis("off")   # remove axis
plt.show()
```

```python
#Rotation using NumPy
rotated = np.rot90(img_array)

rot_img = Image.fromarray(rotated)
rot_img.show()
# Show image
plt.imshow(rot_img)
plt.axis("off")   # remove axis
plt.show()
```
